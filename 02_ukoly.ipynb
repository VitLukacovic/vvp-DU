## 1.4 Funkce, podmínky a cykly

1. Vytvořte funkci, která vrací faktoriál přirozeného čísla.

````python
def faktorial(n):
    vysledek = 1
    for i in range(1, n + 1):
        vysledek *= i
    return vysledek

n = 5
vysledek = faktorial(n)
print(vysledek)
````

2. Stejnou funkci napište i rekurzivně.

````python
def faktorial_rek(n):
    if n > 1:
        return n * faktorial_rek(n - 1)
    else:
        return 1

n = 5
vysledek = faktorial_rek(n)
print(vysledek)
````

3. Vytvořte funkci, která vrací průměr čísel v seznamu.

````python
def prumer(list):
    length = len(list)
    soucet = 0
    if length == 1:
        return list[0]
    
    for i in range(0, length):
        soucet = soucet + list[i]
    return soucet / length

seznam = [9,1,3,4,5,6]
vysledek = prumer(seznam)
print(vysledek)
````

4. Vytvořte funkci, která najde prvních `n` prvočísel.

````python
def prvocisla(n):
    prvocisla_list = [2]
    cislo = 3
    while(len(prvocisla_list) < n):
        is_prvocislo = True
        for i in range(2, int(cislo/2)+1):
            if cislo % i == 0:
                is_prvocislo = False
        if is_prvocislo == True:
            prvocisla_list.append(cislo)
        cislo += 1
    return prvocisla_list

n = 7
vysledek = prvocisla(n)
print(vysledek)
````

# 2. Ladění

**Euklidův algoritmus pro výpočet největšího společného dělitele (NSD) dvou čísel**

```text
Mějme dvě přirozená čísla uložená v proměnných u a w.
Dokud w není 0, opakuj:
    r = u % w
    u = w
    w = r
Po skončení cyklu je v u uložen NSD původních čísel.
```

## 2.1 Úkoly

- Nastavte breakpoint na řádek 3, spusťte buňku v debug režimu a sledujte změny hodnot proměnných `u` a `w` během celého algoritmu.
- Vytvořte podmíněný breakpoint na řádku 5, který se zastaví až ve chvíli, kdy bude `w == 0`.
- Nastavte logovací breakpoint na řádku 5, který vypíše hodnoty proměnných `u` a `w`.


````python
def gcd(a, b):
    # Euklidův algoritmus pro výpočet největšího společného dělitele (NSD) dvou čísel
    u = a
    w = b
    while w != 0:
        r = u % w
        u = w
        w = r
    return u


a = 2 * 2 * 3 * 5 * 7 * 11 * 13
b = 2 * 3 * 3 * 7 * 17

print(a)
print(b)

gcd(a, b)
````
