## 1.1 `ndarray`

1. Jak zjistit velikost paměti libovolného pole? Vyzkoušejte na `a = np.array([1, 2, 3])`. Otestujte různé datové typy (`dtype`), například `np.int32`, `np.float64`, `np.complex128`.


````python
import numpy as np

dtypes = [np.int32, np.float64, np.complex128]

for dt in dtypes:
    a = np.array([1, 2, 3], dtype=dt)
    print(f"dtype: {dt}")
    print(f"  itemsize: {a.itemsize} B")
    print(f"  size: {a.size}")
    print(f"  nbytes: {a.nbytes} B")
    print()
````

2. Vytvořte tři náhodné matice 5x5 a uložte je do souboru ve formátu HDF5.


````python
import h5py

A = np.random.rand(5,5)
B = np.random.rand(5,5)
C = np.random.rand(5,5)

with h5py.File("matice.h5", "w") as f:
    f.create_dataset("matrix_A", data=A)
    f.create_dataset("matrix_B", data=B)
    f.create_dataset("matrix_C", data=C)
````

3. Načtěte soubor z předchozího úkolu, zobrazte obsah každé matice a zjistěte jejich rozměry.


````python
with h5py.File("matice.h5", "r") as f:
    for name in f.keys():
        data = f[name][:]
        print(f"Dataset: {name}")
        print("Obsah matice:\n", data)
        print("Rozměry:", data.shape)
        print("-" * 30)
````
