10. Upravte funkci nalézající prvočísla tak, aby ukládala už spočítaná prvočísla do listu v defaultním argumentu funkce.

- Funkce má uchovávat už spočítaná prvočísla v defaultním argumentu.
- Při dalším volání má dopočítat jen chybějící prvočísla.
- Vracejte vždy **kopii** požadované části seznamu.


````python
def najdi_n_prvocisel(n):
    prvocisla = []
    i = 2
    while len(prvocisla) < n:
        je_prvocislo = True
        for prvocislo in prvocisla:
            if i % prvocislo == 0:
                je_prvocislo = False
                break
        if je_prvocislo:
            prvocisla.append(i)
        i += 1
    return prvocisla

najdi_n_prvocisel(10)
````

````python
def dopln_prvocisla(n, prvocisla=[2]):
    i = prvocisla[-1] + 1

    while len(prvocisla) < n:
        je_prvocislo = True

        for prvocislo in prvocisla:
            if i % prvocislo == 0:
                je_prvocislo = False
                break

        if je_prvocislo:
            prvocisla.append(i)

        i += 1
        
    return prvocisla[:n]

print(dopln_prvocisla(20))
````

````python
class Prvocisla:
    def __init__(self):
        self.prvocisla = [2]

    def __call__(self, n):
        i = self.prvocisla[-1] + 1

        while len(self.prvocisla) < n:
            je_prvocislo = True

            for prvocislo in self.prvocisla:
                if i % prvocislo == 0:
                    je_prvocislo = False
                    break

            if je_prvocislo:
                self.prvocisla.append(i)

            i += 1

        return self.prvocisla[:n]
    
prvocisla = Prvocisla()
prvocisla(3)
````

3. Rozšiřte třídu o metodu `__getitem__`, která bude vracet prvočísla podle indexu (podporujte celočíselné indexy i indexy ve formě `slice`).

- Hodit se vám bude `isinstance` pro rozlišení, zda je vstup `slice` nebo `int`.
- Připomeňte si, že objekt typu `slice` má atributy `start`, `stop` a `step`.


````python
class Prvocisla:
    def __init__(self):
        self.prvocisla = [2]

    def __call__(self, n):
        i = self.prvocisla[-1] + 1

        while len(self.prvocisla) < n:
            je_prvocislo = True

            for prvocislo in self.prvocisla:
                if i % prvocislo == 0:
                    je_prvocislo = False
                    break

            if je_prvocislo:
                self.prvocisla.append(i)

            i += 1

        return self.prvocisla[:n]
    
    def __getitem__(self, n):
        if isinstance(n, int):
            if n < 0:
                raise IndexError("Záporné indexy nejsou podporovány")

            self(n + 1)
            return self.prvocisla[n]

        elif isinstance(n, slice):
            start = n.start or 0
            stop = n.stop
            step = n.step or 1

            if stop is None:
                raise ValueError("Slice musí mít definovaný 'stop'")

            self(stop)
            return self.prvocisla[start:stop:step]
````

```python
class Zlomek:
    def __init__(self, citatel, jmenovatel):
    def __add__(self, other):
    def __sub__(self, other):
    def __mul__(self, other):
    def __truediv__(self, other):
    def simplify(self):
    def __str__(self):
    def __repr__(self):
    def to_float(self):
    def __le__(self, other):
```

````python
import math

class Zlomek:
    def __init__(self, citatel, jmenovatel):
        if jmenovatel == 0:
            raise ValueError("Jmenovatel nemůže být nula")
        
        self.citatel = citatel
        self.jmenovatel = jmenovatel
        
    def __add__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        cit = self.citatel * other.jmenovatel + other.citatel * self.jmenovatel
        jmen = self.jmenovatel * other.jmenovatel
        return Zlomek(cit, jmen)
    
    def __sub__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        cit = self.citatel * other.jmenovatel - other.citatel * self.jmenovatel
        jmen = self.jmenovatel * other.jmenovatel
        return Zlomek(cit, jmen)
    
    def __mul__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        return Zlomek(self.citatel * other.citatel, self.jmenovatel * other.jmenovatel)
    
    def __truediv__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        if other.citatel == 0:
            raise ZeroDivisionError("Dělení nulou")
        return Zlomek(self.citatel * other.jmenovatel, self.jmenovatel * other.citatel)

    def simplify(self):
        gcd = math.gcd(self.citatel, self.jmenovatel)
        self.citatel //= gcd
        self.jmenovatel //= gcd
        if self.jmenovatel < 0:
            self.citatel *= -1
            self.jmenovatel *= -1

    def __str__(self):
        return f"{self.citatel}/{self.jmenovatel}"

    def __repr__(self):
        return f"Zlomek({self.citatel}, {self.jmenovatel})"

    def to_float(self):
        return self.citatel / self.jmenovatel
    
    def __le__(self, other):
        return self < other or self == other
    
    def __lt__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        return self.citatel * other.jmenovatel < self.jmenovatel * other.citatel
    
    def __ge__(self, other):
        return not self < other
    
    def __gt__(self, other):
        return not self <= other

    def __eq__(self, other):
        if not isinstance(other, Zlomek):
            return NotImplemented
        return self.citatel * other.jmenovatel == self.jmenovatel * other.citatel

    

    
````
