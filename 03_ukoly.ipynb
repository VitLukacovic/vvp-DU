## 1.6 For cyklus s `else`
Vypište první tři prvky seznamu. Pokud je seznam dlouhý nejvýše 3 prvky, vypište na závěr `"Krátký seznam!"`.

Vyzkoušejte na `[1, 2, 3, 4, 5]` a `[1, 2]`.


````python
def prvni_tri_prvky(sz):
    if len(sz) < 3:
        print("Krátký seznam!")
    else:
        for i in range(0, 3):
            print(sz[i])

seznam1 = [1,2,3,4,5]
seznam2 = [1,2]           
prvni_tri_prvky(seznam1)
    
````

## 1.7 For cyklus s `continue`
Vypište všechna sudá čísla ze seznamu `[1, 2, 3, 4, 5]`.


````python
seznam1 = [1,2,3,4,5]
delka = len(seznam1)

for i in range(0, delka):
    if seznam1[i] % 2 == 0:
        print(seznam1[i])
    else:
        continue
````

````python
def uroceni(pocatecni_vklad, urokova_sazba, rocni_vklad, finalni_castka):
    pocet_let = 0
    aktualni_castka = pocatecni_vklad
    while aktualni_castka <= finalni_castka:
        aktualni_castka = aktualni_castka * (1 + urokova_sazba)
        pocet_let += 1
        aktualni_castka += rocni_vklad
    print("počet let:", pocet_let, ", částka:", aktualni_castka)

uroceni(10000, 0.09, 1000, 1000000)
        
````

# 2. Generátory a iterátory
## 2.1 Fibonacciho generátor
Napište generátor (funkci), který vytváří Fibonacciho posloupnost a postupně vrátí prvních `n` členů. Otestujte pomocí `for` cyklu.


````python
def fibonacci(n):
    a, b = 1, 1
    for i in range(n):
        yield a
        a, b = b, a + b
        
for i in fibonacci(10):
    print(i)
````

## 2.2 Generátorový výraz
Napište generátorový výraz, který vrací všechna sudá čísla menší než 100. Otestujte pomocí `for` cyklu.


````python
suda_cisla = (i for i in range(100) if i % 2 == 0)

for cislo in suda_cisla:
    print(cislo)
````

## 2.3 Test dělitelnosti pomocí generátoru
Pro dané číslo `n` (berte `n >= 2`) vytvořte generátor, který pro každé `d` od `2` do `n-1` vrací výraz `n % d != 0`. Otestujte pro několik čísel a použijte `all()` tak, aby výraz fungoval jako test prvočíselnosti.


````python
n = 7
        
delitelnost = (n % d != 0 for d in range(2, n))
  
print(all(delitelnost))
    
````

## 3.3 Prvočísla do 100
Vytvořte seznam prvočísel od 1 do 100 pomocí list comprehension (použijte test prvočíselnosti z předchozích úloh).


````python
def je_prvocislo(n):
    return all(n % d != 0 for d in range(2, n))

print([i for i in range(1, 101) if je_prvocislo(i)])
````

# 4. Jupyter magics
## 4.1 Vytvoření modulu přes `%%writefile`
Použijte `%%writefile` a vytvořte soubor `factorial.py` se dvěma funkcemi:

```python
def factorial_loop(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
```

Modul načtěte a zavolejte obě funkce.


````python
%%writefile factorial.py

def factorial_loop(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
````

## 4.2 Měření času
Použijte `%time`, `%timeit`, `%%time` a `%%timeit` k porovnání času běhu dvou variant výpočtu faktoriálu (např. pro `n = 200`).


````python
%run factorial.py

%time factorial_loop(200)
print()
%time factorial_recursive(200)
````
